```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
data = pd.read_csv('Predict Academic Success - data.csv')
```

```python
# Get the target and feature of the dataset
X = data.drop(columns=["Target"]) # Features
y = data["Target"] # Target
```

```python
# Create and train the Decision Tree model
dtree = DecisionTreeClassifier(max_depth=3)
dtree.fit(X, y)

# Plot Decision Tree
fig, ax = plt.subplots(figsize=(15, 10), dpi=300) 
tree.plot_tree(dtree, feature_names=X.columns, filled=True, class_names=["Dropout", "Enrolled", "Graduate"], ax=ax)
fig.savefig("decision_tree.png", dpi=400, bbox_inches="tight")


# Show the image
plt.show()
```
